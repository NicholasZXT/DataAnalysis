检查分类变量的列取值都有哪些以及取值个数

```python
# str_columns = ['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService','OnlineSecurity','DeviceProtection'
#                ,'TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling', 'PaymentMethod']
# 选择需要进行哑编码的列
columns = list(data.columns)
#  注意，那些本来就是数值的列如果不排除的话，也会被get_dummies当做分类变量看待，所以也需要排除
columns_dummy = [ele for ele in columns if ele not in ['customerID', 'SeniorCitizen','tenure','MonthlyCharges','TotalCharges','Churn']]

# data[columns_dummy].apply(pd.Series.value_counts).fillna("")
data[columns_dummy].apply(pd.Series.value_counts).count()
```

```python
# data_dummy = pd.get_dummies(data, columns=columns_dummy)
data_dummy = pd.get_dummies(data, columns=columns_dummy, drop_first=True)
```
